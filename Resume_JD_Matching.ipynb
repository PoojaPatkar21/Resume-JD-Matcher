```python
import PyPDF2
import os

# Specify the directory containing the PDF files-
pdf_directory = 'drive/My Drive/DATA/data/data/INFORMATION-TECHNOLOGY/'

# Initialize empty lists to store extracted data
categories = []
skills = []
educations = []

# Iterate through PDF files in the directory
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):
        pdf_file = os.path.join(pdf_directory, filename)

        # Initialize variables to store data from the current PDF
        category = ""
        skills_text = ""
        education = ""

        # Open and read the PDF file
        with open(pdf_file, 'rb') as pdf:
            pdf_reader = PyPDF2.PdfReader(pdf)

            # Iterate through PDF pages
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                page_text = page.extract_text()

                # Implement logic to extract Category, Skills, and Education

                if "Objective:" in page_text:
                    category = page_text.split("Objective:")[1].strip()


                if "Skills:" in page_text:
                    skills_text = page_text.split("Skills:")[1].strip()
                    # Split the skills text into a list
                    skills = [s.strip() for s in skills_text.split(",")]


                if "Education:" in page_text:
                    education_text = page_text.split("Education:")[1].strip()


        # Append extracted data to respective lists
        categories.append(category)
        skills.append(skills_text)
        educations.append(education)

        # Print information for each file
        print("Filename:", filename)
        print("Objective:", category)
        print("Skills:", ", ".join(skills))
        print("Education:", education)
        print("------")


```

#  Candidate-Job Matching

```python
!pip install transformers
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
from sklearn.metrics.pairwise import cosine_similarity

# Define file paths to job description and CV text files
job_description_file = "/content/ITJD.txt"
cv_text_file = "/content/Resume.txt"

# Read text from files
with open(job_description_file, "r", encoding="utf-8") as job_desc_file:
    job_description = job_desc_file.read()

with open(cv_text_file, "r", encoding="utf-8") as cv_text_file:
    cv_text = cv_text_file.read()

# Print the job description and CV text
print("Job Description:")
print(job_description)
print("\nCV Text:")
print(cv_text)






```

Tokenization

```python
# Tokenization
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
job_description_tokens = tokenizer(job_description, return_tensors="pt", padding=True, truncation=True)
cv_tokens = tokenizer(cv_text, return_tensors="pt", padding=True, truncation=True)

# Print tokenized input
print("\nTokenized Job Description:")
print(tokenizer.decode(job_description_tokens['input_ids'][0]))
print("\nTokenized CV Text:")
print(tokenizer.decode(cv_tokens['input_ids'][0]))
```

Embedding

```python
# Embedding
model = DistilBertModel.from_pretrained("distilbert-base-uncased")
job_description_embeddings = model(**job_description_tokens).last_hidden_state.mean(dim=1)
cv_embeddings = model(**cv_tokens).last_hidden_state.mean(dim=1)

# Convert tensors to NumPy arrays
job_description_embeddings = job_description_embeddings.detach().numpy()
cv_embeddings = cv_embeddings.detach().numpy()

# Print embeddings (for demonstration, as embeddings can be large)
print("\nJob Description Embeddings:")
print(job_description_embeddings)
print("\nCV Embeddings:")
print(cv_embeddings)
```

Cosine similarity

```python
# Calculate cosine similarity
similarity_scores = cosine_similarity(job_description_embeddings, cv_embeddings)

# Print similarity scores
print("\nCosine Similarity Scores:")
print(similarity_scores)
```

# Ranking Candidates- Top 5

```python
# Rank CVs based on similarity scores
top_5_candidates_indices = similarity_scores.argsort()[0][-5:][::-1]

# Print top 5 candidate indices
print("\nTop 5 Candidate Indices:")
print(top_5_candidates_indices)
```
